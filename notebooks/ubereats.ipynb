```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.ubereats.com/ch/category/genève-ge"
response=requests.get(url)

soup=BeautifulSoup(response.content, "html.parser")
#print(response.content)

ubereats=soup.find("main", id="main-content")
print(ubereats)
```
