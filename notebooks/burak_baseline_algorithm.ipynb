# Competition

```python
restaurant_type = 'Lebanese'
df_scoring = pd.DataFrame()
```

```python
# total competition
df_scoring['total_competition'] = df_rest[['combined_main_category', 'cluster']]\
    .groupby('cluster').count()
```

```python
# direct competition
df_scoring['direct_competition'] = df_rest[['combined_main_category', 'cluster']]\
    [df_rest['combined_main_category'].str.contains(restaurant_type, na = False)]\
    .groupby('cluster').count()
```

```python
# total satisfaction
df_scoring['total_satisfaction'] = df_rest[['combined_main_category', 'cluster', 'rating']]\
    .groupby('cluster').mean()
```

```python
# direct satisfaction
df_scoring['direct_satisfaction'] = df_rest[['combined_main_category', 'cluster', 'rating']]\
    [df_rest['combined_main_category'].str.contains(restaurant_type, na = False)]\
    .groupby('cluster').mean()
```

```python
df_scoring = df_scoring.reset_index()
df_scoring
```

```python
from sklearn.preprocessing import MinMaxScaler

mm_scaler = MinMaxScaler()
cols = df_scoring.drop(columns='cluster')
mm_scaler.fit(cols)
df_scoring_norm = pd.DataFrame(mm_scaler.transform(cols), columns=cols.columns)
df_scoring_norm
```

```python
df_scoring_norm['cluster'] = df_scoring['cluster']
```

```python
df_scoring_norm['final_score'] = 0.25 * (1-df_scoring_norm['total_competition'])\
    + 0.25 * (1-df_scoring_norm['direct_competition'])\
    + 0.25 * (1-df_scoring_norm['total_satisfaction'])\
    + 0.25 * (1-df_scoring_norm['direct_satisfaction'])
```
