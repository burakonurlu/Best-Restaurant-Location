```python
# required module import
import shapely.geometry as shg
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
# reading GVA district shape data and converting it to lat / lng format
df_district_shp = gpd.read_file('../raw_data/VDG_QUARTIER_VILLE.shp')
display(df_district_shp)
df_district_geo = df_district_shp.to_crs(epsg=4326)
display(df_district_geo)
```

```python
# reading lat / lng from df and converting it to numpy array
df_rest = pd.read_csv('../data/data_combined_v1.01.csv').drop(columns=["Unnamed: 0"])
df_rest.head()
```

# Initial Mapping

```python
for idx_rest, row_rest in df_rest.iterrows():
    point = shg.Point(df_rest.loc[idx_rest,'geometry.location.lng'],\
                 df_rest.loc[idx_rest,'geometry.location.lat'])
    for idx_geo, row_geo in df_district_geo.iterrows():
        if point.within(df_district_geo.loc[idx_geo, 'geometry']):
            df_rest.loc[idx_rest, 'district'] = df_district_geo.loc[idx_geo, 'NOM_QUARTI']
            break
```

```python
df_rest['district'].value_counts()
```

```python
df_rest['district'].isnull().sum()
```

```python
# Sample Check 1
pointx = shg.Point((6.1561736, 46.1993241)) # Luigia in Eaux-Vives area
print(pointx.within(df_district_geo['geometry'][3])) # Plainpalais area
print(pointx.within(df_district_geo['geometry'][4])) # Eaux-Vives area
```

```python
# Sample Check 2
pointx = shg.Point((6.1361528, 46.1980208)) # Umamido in Plainpalais area
print(pointx.within(df_district_geo['geometry'][3])) # Plainpalais area
print(pointx.within(df_district_geo['geometry'][4])) # Eaux-Vives area
```

```python
df_rest.columns
```

```python
# Final Processing
# Removal of Future Restaurants i.e. without restaurant category
df_rest = df_rest[(df_rest['combined_main_category'].notnull()) &\
                  (df_rest['district'].notnull())]\
                .drop(columns=["Unnamed: 0"])\
                .reset_index(drop=True)
df_rest
```

```python
df_rest.to_csv('../data/data_combined_v1.02.csv', encoding='utf-8-sig', index=False)
```
